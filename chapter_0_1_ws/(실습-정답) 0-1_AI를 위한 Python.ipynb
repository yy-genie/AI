## Step 1: Python 기본 자료형

### Concept Check: Python의 기본 자료형

Python에서 데이터를 다루는 가장 기본적인 단위는 **자료형(Data Type)**입니다. 자료형은 데이터의 종류를 나타내며, 각 자료형마다 사용할 수 있는 연산과 메서드가 다릅니다.

**주요 기본 자료형**:
- **int (정수형)**: 정수 값을 저장합니다. 예: `10`, `-5`, `0`
- **float (부동소수점형)**: 소수점을 포함한 숫자를 저장합니다. 예: `3.14`, `-0.5`
- **str (문자열형)**: 문자들의 집합을 저장합니다. 예: `"Hello"`, `'Python'`
- **bool (불리언형)**: 참(True) 또는 거짓(False)을 나타냅니다.
- **None**: 값이 없음을 명시적으로 표현합니다. (Java의 null과 유사)

**왜 자료형이 중요한가?**
- 각 자료형은 서로 다른 연산을 지원합니다. 예를 들어, 숫자는 사칙연산이 가능하지만 문자열은 불가능합니다.
- 자료형을 이해하면 데이터를 올바르게 처리하고, 에러를 예방할 수 있습니다.

**학습 목표**: Python에서 사용하는 기본 자료형 (`int`, `float`, `str`, `list`, `dict`, `set`)을 이해하고, 조건문/반복문, 함수를 사용할 수 있습니다.

**진행하는 실습 요약**: 기초적인 자료형의 인스턴스화 방법과 내장된 메소드를 직접 사용해봅니다.

## Step 2: 컬렉션 자료형

### Concept Check: 컬렉션 자료형이란? 왜 필요한가?

**컬렉션 자료형이 필요한 이유**
실제 프로그래밍에서는 하나의 데이터가 아니라 여러 개의 데이터를 함께 다루는 경우가 많습니다:
- 학생 10명의 점수를 저장하고 싶다면?
- 장바구니에 담긴 상품 목록을 관리하려면?
- 학생 이름과 점수를 함께 저장하려면?
이럴 때 컬렉션 자료형을 사용합니다!

**컬렉션 자료형이란?**
- 여러 개의 데이터를 하나의 변수에 저장할 수 있는 자료형입니다.
- Python에는 4가지 주요 컬렉션 자료형이 있습니다: list, tuple, set, dict
- 각 자료형은 데이터를 저장하는 방식과 용도가 다릅니다.

📋 List (리스트)

Python에는 `dict`에 부가적인 기능을 탑재한 빌트인 딕셔너리가 추가로 존재합니다. 자주 사용되는 두 가지 `dict`가 있는데 바로 `defaultdict`와 `Counter`입니다.
- **`defaultdict`**: 기본값을 지정할 수 있는 딕셔너리로,  키가 존재하지 않을 때 KeyError 대신 지정된 기본값을 자동으로 생성합니다.
- **`Counter`**: 리스트나 문자열 등 반복 가능한 객체의 요소 개수를 자동으로 세어주는 딕셔너리로 가장 많이 등장한 요소를 쉽게 파악할 수 있습니다.


````python
from collections import defaultdict, Counter

# defaultdict 예시
dd = defaultdict(int)
# 아직 `apple`이라는 key값에 해당하는 value를 만들어주지 않았으나
# += 1 연산이 가능함.
dd['apple'] += 1
dd['banana'] += 2
print("defaultdict 결과:", dict(dd))  # {'apple': 1, 'banana': 2}

# Counter 예시
fruits = ['apple', 'banana', 'apple', 'cheese', 'banana', 'banana']
counter = Counter(fruits)
print("Counter 결과:", dict(counter))  # {'apple': 2, 'banana': 3, 'cheese': 1}
````

## Step 3: 제어문과 반복문

### Concept Check: 제어문과 반복문이란? 왜 필요한가?
**제어문과 반복문이 필요한 이유**
프로그램은 단순히 순차적으로 실행되는 것이 아니라:
- 조건에 따라 다르게 동작해야 할 때가 있습니다 (예: 로그인 성공/실패)
- 같은 작업을 반복해야 할 때가 있습니다 (예: 100명의 성적 처리)
이럴 때 제어문과 반복문을 사용합니다!

**제어문과 반복문이란?**
- 제어문 (Control Flow Statement): 프로그램의 실행 흐름을 제어하는 문장
- 조건문 (if, elif, else): 조건에 따라 다른 코드 실행
- 반복문 (for, while): 같은 코드를 여러 번 실행

````python
# 1. 조건문 (if, elif, else) - 상황에 따라 다르게 동작
score = 85

if score >= 90:
    print("학점: A")
elif score >= 80:
    print("학점: B")
elif score >= 70:
    print("학점: C")
else:
    print("학점: D")

# 논리 연산자 사용 예시
age = 20
has_card = True

if age >= 19 and has_card:
    print("입장 가능합니다.")
else:
    print("입장 불가능합니다.")
````

````python
# 2. 반복문 (for) - 정해진 횟수만큼 반복
# for 루프: 순회 가능한(iterable) 객체의 요소를 하나씩 순회합니다.
# Java의 enhanced for loop (for-each)와 유사합니다.
fruits = ["apple", "banana", "cherry"]
print("\n--- 과일 목록 ---")
for fruit in fruits:
    print(fruit)
````

````python
# range() 함수와 함께 사용하는 for 루프: 특정 횟수만큼 반복할 때 유용합니다.
# Java의 for (int i = 0; i < n; i++)와 유사합니다.
print("\n--- 0부터 4까지 출력 ---")
for i in range(5): # 0부터 4까지 (5는 포함 안됨)
    print(i)

print("\n--- 2부터 10까지 짝수 출력 ---")
for i in range(2, 11, 2): # 시작 값, 끝 값 (포함 안됨), 스텝
    print(i)
````

````python
# 3. 반복문 (while) - 조건이 참인 동안 반복
# while 루프: 조건이 True인 동안 반복합니다.
print("\n--- 카운트 다운 ---")
count = 5
while count > 0:
    print(count)
    count -= 1
print("발사!")
````

````python
# break와 continue
print("\n--- break와 continue ---")
for i in range(10):
    if i == 3:
        continue # 현재 반복 건너뛰고 다음 반복으로
    if i == 7:
        break # 루프 완전히 종료
    print(i)
````

리스트 컴프리헨션 (List Comprehension)
- 반복문을 간결하게 표현하는 Python의 강력한 기능
- Java의 Stream API의 map, filter와 유사한 개념으로 볼 수 있습니다.
- 🌟굉장히 아주 자주 매우🌟 등장하기 때문에 익숙해져야합니다.
- https://docs.python.org/3/tutorial/datastructures.html#list-comprehensions

````python
print("\n--- 리스트 컴프리헨션 ---")
squares = [x**2 for x in range(10)]
print(f"0부터 9까지 제곱: {squares}")

even_numbers = [x for x in range(20) if x % 2 == 0]
print(f"0부터 19까지 짝수: {even_numbers}")
````

- `product()`:	데카르트 곱 (중첩 반복)
- `permutations()`:	순열 생성
- `combinations()`:	조합 생성

````python
from itertools import product, permutations, combinations, accumulate

# product: 모든 조합 (중첩 반복)
print("\n--- product ---")
for p in product([1, 2], ['A', 'B']):
    print(p)

# permutations: 순열
print("\n--- permutations ---")
for p in permutations([1, 2, 3], 2):
    print(p)

# combinations: 조합
print("\n--- combinations ---")
for c in combinations([1, 2, 3], 2):
    print(c)

# accumulate: 누적합
print("\n--- accumulate ---")
import operator
nums = [1, 2, 3, 4]
print("누적합:", list(accumulate(nums)))
print("누적곱:", list(accumulate(nums, func=operator.mul)))
````

## Step 5: Python을 이용한 객체지향프로그래밍

### Concept Check: 클래스(Class)란 무엇인가?

지금까지 우리는 Python의 기본 자료형(int, str, list 등)과 함수를 배웠습니다. 이제 한 단계 더 나아가 **클래스(Class)**를 배워봅시다.

**클래스란?**
- 클래스는 **데이터(속성)와 기능(메서드)을 하나로 묶는 틀**입니다.
- 마치 "붕어빵 틀"과 같습니다. 틀(클래스)을 만들면, 그 틀로 여러 개의 붕어빵(객체)을 찍어낼 수 있습니다.

**왜 클래스를 사용하나요?**
1. **구조화**: 관련된 데이터와 기능을 하나로 묶어 코드를 체계적으로 관리할 수 있습니다.
2. **재사용성**: 한 번 클래스를 만들면, 여러 번 재사용할 수 있습니다.
3. **유지보수**: 코드 수정이 필요할 때 클래스만 수정하면 되므로 유지보수가 쉽습니다.

**핵심 용어 정리**
- **클래스 (Class)**: 객체를 만들기 위한 틀 (예: Dog 클래스)
- **객체 (Object) / 인스턴스 (Instance)**: 클래스로부터 만들어진 실제 데이터 (예: my_dog = Dog())
  - 객체와 인스턴스는 같은 의미로 사용됩니다. "my_dog는 Dog 클래스의 인스턴스다" = "my_dog는 객체다"
- **속성 (Attribute)**: 객체가 가지는 데이터 (예: my_dog.name, my_dog.breed)
- **메서드 (Method)**: 객체가 수행할 수 있는 기능 (예: my_dog.bark())
- **생성자 (__init__)**: 객체가 생성될 때 자동으로 호출되는 특별한 메서드로, 객체의 초기 상태를 설정합니다.

**클래스를 사용하는 이유를 예시로 이해하기**

강아지 정보를 관리한다고 가정해봅시다.

**❌ 클래스 없이 (비효율적)**
```python
dog1_name = "바둑이"
dog1_breed = "진돗개"
dog2_name = "초코"
dog2_breed = "푸들"

# 강아지가 짖는 기능을 만들려면?
def bark_dog1():
    print(f"{dog1_name}가 멍멍 짖습니다!")
    
def bark_dog2():
    print(f"{dog2_name}가 멍멍 짖습니다!")
```
→ 강아지가 100마리면? 변수와 함수를 200개나 만들어야 합니다!

**✅ 클래스 사용 (효율적)**
```python
class Dog:
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed
    
    def bark(self):
        print(f"{self.name}가 멍멍 짖습니다!")

# 여러 강아지를 쉽게 만들 수 있습니다
dog1 = Dog("바둑이", "진돗개")
dog2 = Dog("초코", "푸들")
dog1.bark()  # 바둑이가 멍멍 짖습니다!
dog2.bark()  # 초코가 멍멍 짖습니다!
```
→ 강아지가 100마리여도 간단하게 관리 가능!


````python
# 클래스 정의
class Dog:
  # 생성자 (Java의 생성자와 유사)
  def __init__(self, name, breed):
    self.name = name # 인스턴스 변수 (속성)
    self.breed = breed

  # 메서드 (Java의 메서드와 유사)
  def bark(self):
    print(f"{self.name}가 멍멍 짖습니다!")

  def __str__(self):
    # 객체를 문자열로 표현할 때 사용 (Java의 toString()과 유사)
    return f"이름: {self.name}, 품종: {self.breed}"

# 객체 생성
my_dog = Dog(name="바둑이", breed="진돗개")

# 속성 접근
print(f"우리 강아지 이름: {my_dog.name}")
print(f"우리 강아지 품종: {my_dog.breed}")

# 메서드 호출
my_dog.bark()

# 객체 출력 (toString()과 유사)
print(my_dog)
````

````python
# 상속 (Inheritance)
class Puppy(Dog):
  def __init__(self, name, breed, age):
    # 부모 클래스의 생성자 호출 (Java의 super()와 유사)
    super().__init__(name, breed)
    self.age = age

  # 메서드 오버라이딩 (Method Overriding)
  def bark(self):
    print(f"{self.name}가 낑낑거립니다!")

  def __str__(self):
    return f"이름: {self.name}, 품종: {self.breed}, 나이: {self.age}개월"

# 자식 클래스 객체 생성
my_puppy = Puppy(name="초코", breed="푸들", age=3)

# 상속된 메서드 및 오버라이딩된 메서드 호출
my_puppy.bark()
print(my_puppy)
````

````python
# 캡슐화 (Encapsulation) - Python은 명시적인 private 키워드는 없지만, 관례적으로 변수 이름 앞에 `_` 또는 `__`를 붙여 내부적으로 사용함을 나타냅니다.
class BankAccount:
  def __init__(self, balance):
    self.__balance = balance # private 변수 (관례)

  def deposit(self, amount):
    if amount > 0:
      self.__balance += amount
      print(f"{amount}원 입금되었습니다. 현재 잔액: {self.__balance}")
    else:
      print("유효하지 않은 금액입니다.")

  def withdraw(self, amount):
    if amount > 0 and amount <= self.__balance:
      self.__balance -= amount
      print(f"{amount}원 출금되었습니다. 현재 잔액: {self.__balance}")
    else:
      print("잔액이 부족하거나 유효하지 않은 금액입니다.")

  def get_balance(self):
    return self.__balance

# 캡슐화 예시
account = BankAccount(balance=10000)
account.deposit(amount=5000)
account.withdraw(amount=2000)
# print(account.__balance) # 직접 접근은 어렵습니다 (Name mangling)
print(f"현재 잔액: {account.get_balance()}")
````

### TODO 8: `class` 만들어보기

**요구사항**
1. 다음과 같은 속성과 메소드를 포함하는 `Student` 클래스를 만들어주세요.
   - 이름 (`name: str`)
   - 국어 점수 (`korean: int`)
   - 영어 점수 (`english: int`)
   - 수학 점수 (`math: int`)
   - `get_average()` 메서드를 통해 평균 점수를 계산할 수 있습니다.
2. 함수: `check_pass(student: Student) -> str`
   - 	평균 점수가 75점 이상이면 "합격", 아니면 "불합격" 반환
3. 함수: `analyze_students(students: list[Student], threshold: int) -> tuple[list[str], list[str]]`
   - 모든 학생의 합격/불합격 여부를 분석
   - 총점이 threshold점 이상인 학생을 우수 학생으로 선정
   - 반환값: (합격자 이름 리스트, 우수 학생 이름 리스트)


````python
from typing import List, Tuple


# None 부분을 직접 채우며 테스트코드가 동작하도록 문제를 풀어보세요 :)
# 정답을 확인하는 assert 코드는 수정하지말고 진행해주세요
class Student:
    def __init__(self, name: str, korean: int, english: int, math: int) -> None:
        # 학생 이름과 과목별 점수를 저장합니다.
        self.name = name
        self.korean = korean
        self.english = english
        self.math = math

    def get_total(self) -> int:
        # 총점을 계산하는 메서드입니다.
        return self.korean + self.english + self.math

    def get_average(self) -> float:
        # 평균 점수를 계산하는 메서드입니다.
        return self.get_total() / 3


def check_pass(student: Student) -> str:
    # 학생 평균 점수가 75점 이상이면 "합격", 아니면 "불합격"을 반환합니다.
    if student.get_average() >= 75:
        return "합격"
    else:
        return "불합격"

    return "합격" if student.get_average() >= 75 else "불합격"


def analyze_students(
    students: List[Student], threshold: int
) -> Tuple[List[str], List[str]]:
    # 여러 학생 정보를 받아 합격자와 우수자를 분류합니다.
    # threshold 이상 총점을 받은 학생은 우수자로 간주합니다.
    passed = []
    excellent = []

    for student in students:
        # student 가 합격, 우수자인지 판별
        if check_pass(student) == "합격":
            passed.append(student.name)
        if student.get_total() >= threshold:
            excellent.append(student.name)

    return passed, excellent


# 아래는 채점코드입니다.
# 수정하지말고 진행해주세요 :)
students: List[Student] = [
    Student(name="뷔", korean=90, english=95, math=100),  # 평균 95.0, 총점 285
    Student(name="정국", korean=60, english=70, math=65),  # 평균 65.0, 총점 195
    Student(name="지민", korean=80, english=90, math=85),  # 평균 85.0, 총점 255
    Student(name="슈가", korean=75, english=70, math=80),  # 평균 75.0, 총점 225
]

passed, excellent = analyze_students(students, threshold=250)

assert set(passed) == {"지민", "뷔", "슈가"}, "합격자 리스트 오류"
assert set(excellent) == {"지민", "뷔"}, "우수 학생 리스트 오류"
assert check_pass(students[1]) == "불합격", "불합격자 판별 오류"
assert check_pass(students[2]) == "합격", "합격자 판별 오류"

print("✅ 모든 테스트를 통과했습니다!")
````

### 📝 **참고**: 사실 모든 것은 객체다

Python의 모든 기본 자료형(int, str, list, dict 등)은 사실 클래스(class) 로 정의되어 있고, 그 내부에 정의된 특수 메서드들(소위 매직 메서드 또는 덧셈 연산자 오버로딩용 메서드)을 통해 우리가 사용하는 연산자나 함수가 동작하는 것입니다.

```python
print(type(10))        # <class 'int'>
print(type("hello"))   # <class 'str'>
print(type([1, 2, 3]))  # <class 'list'>
```

이처럼 int, str, list, dict, float 등은 전부 class로 정의되어 있으며, 우리가 사용하는 수많은 기능들은 그 내부의 특수한 메서드들을 통해 구현되어 있습니다.

🪄 Magic Method란?

매직 메서드(Magic Method) 또는 덧셈 연산자 오버로딩용 메서드는 이름 앞뒤에 __ (double underscore)를 붙인 Python 메서드들입니다. 예를 들어:
-	`__init__`: 생성자 (객체가 생성될 때 호출됨)
-	`__str__`: 문자열로 표현할 때 사용 (print 등)
-	`__add__`: + 연산자 동작을 정의
-	`__len__`: len() 함수 호출 시 동작
-	`__getitem__`: 인덱싱 (`obj[key]`|)
-	`__eq__`: == 연산자 오버라이딩

이런 매직 메서드들은 Python의 핵심 문법을 클래스 수준에서 재정의하거나 확장할 수 있게 해줍니다.

예시:
```python
a = 10
b = 5
print(a + b)            # 15
print(a.__add__(b))     # 15, 동일한 결과
```

Custom Class에 적용해보기
```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

v1 = Vector(1, 2)
v2 = Vector(3, 4)
print(v1 + v2)  # (4, 6)
```

이처럼 사실 모든 자료형 또한 객체라는 것을 알았습니다. 우리는 앞으로 python을 다양하게 활용하다보면 magic method를 생각보다 직접 짜서 사용하게 되는 경우들이 있습니다. 참고하면 좋은 [자료](https://docs.python.org/3/reference/datamodel.html#basic-customization)

```python

for num in numbers:
```

**NumPy 배열 vs Python 리스트**

```python
# Python 리스트: 느림, 반복문 필요
numbers = [1, 2, 3, 4, 5]
squared = [x**2 for x in numbers]  # [1, 4, 9, 16, 25]
```

**NumPy 배열: 빠름, 한 줄로 처리**
```python
arr = np.array([1, 2, 3, 4, 5])
squared = arr ** 2  # array([1, 4, 9, 16, 25])
```

- 숫자 데이터를 빠르게 계산하는 도구

- 머신러닝 알고리즘의 수학적 계산을 담당


**클래스와의 연결**: NumPy 배열도 클래스의 인스턴스입니다
```python
arr = np.array([1, 2, 3])
```

**DataFrame 데이터 접근하기**

**주요 데이터 접근 방법**
- `df["컬럼명"]` 또는 `df.컬럼명`: **특정 열(Series) 반환** - 한 개 열의 모든 데이터를 가져올 때 사용
- `df.iloc[행번호]`: **정수 위치로 행 접근** - 0부터 시작하는 숫자로 행을 선택할 때 사용
- `df.loc[인덱스]`: **인덱스 라벨로 행 접근** - 인덱스 이름(라벨)으로 행을 선택할 때 사용
- `df[["컬럼1", "컬럼2"]]`: **여러 열 선택 (DataFrame 반환)** - 복수 열을 동시에 가져올 때 사용

**언제 어떤 방법을 사용하나요?**
- 한 개 열만 필요할 때 → `df["Name"]` (Series 반환)
- 여러 열이 필요할 때 → `df[["Name", "Age"]]` (DataFrame 반환)
- 특정 행을 숫자로 선택할 때 → `df.iloc[0]` (첫 번째 행)
- 특정 행을 인덱스 이름으로 선택할 때 → `df.loc["row1"]`


````python
# Column 접근
print("\nName 열:\n", df["Name"])

# Row 접근 by index (정수 위치 기반)
print("\n1번째 행 (iloc):\n", df.iloc[0])  # Alice

# Row 접근 by label (index label 기반이며, 지금은 숫자와 같습니다)
print("\n1번째 행 (loc):\n", df.loc[0])

# 여러 행/열 추출
print("\nName과 Score 열만 추출:\n", df[["Name", "Score"]])
````

### Guided Build: DataFrame 기본 함수 - 무엇을 반환하는가?

Pandas를 처음 사용할 때 가장 헷갈리는 부분은 **"이 함수가 무엇을 반환하는가?"**입니다. 아래 표를 참고하여 각 함수의 역할을 명확히 이해합시다.

**데이터 탐색 함수**
- `df.info()` → **None을 반환합니다** (출력만 합니다) - DataFrame의 전체 정보(컬럼, 데이터 타입, 메모리 등)를 **화면에 출력**할 때 사용
- `df.describe()` → **DataFrame 반환** - 수치형 컬럼의 통계 요약(평균, 표준편차, 최소/최대값 등)을 **표로 반환**할 때 사용
- `df.shape` → **(행, 열) 튜플 반환** - DataFrame의 크기(행 개수, 열 개수)를 **확인**할 때 사용
- `df.columns` → **Index 객체 반환** - 모든 컬럼 이름을 **리스트처럼 확인**할 때 사용 (tolist()로 리스트 변환 가능)
- `df.index` → **RangeIndex/Index 객체 반환** - 행 인덱스를 **확인**할 때 사용
- `df.head(n)` → **DataFrame 반환** - 상위 n개 행을 **미리보기**할 때 사용 (기본값 n=5)
- `df.tail(n)` → **DataFrame 반환** - 하위 n개 행을 **미리보기**할 때 사용 (기본값 n=5)

**데이터 처리 함수**
- `df["컬럼"].mean()` → **float 반환** - 특정 컬럼의 **평균값을 계산**할 때 사용
- `df["컬럼"].sum()` → **float 반환** - 특정 컬럼의 **합계를 계산**할 때 사용
- `df["컬럼"].max()` → **값 반환** - 특정 컬럼의 **최대값을 찾을** 때 사용
- `df["컬럼"].min()` → **값 반환** - 특정 컬럼의 **최소값을 찾을** 때 사용
- `df.sort_values(by="컬럼")` → **DataFrame 반환** - 특정 컬럼을 기준으로 **정렬된 새 DataFrame**을 만들 때 사용
  - 중요: 원본 DataFrame은 변경되지 않습니다! (inplace=True 옵션을 주면 원본 변경)
- `df.groupby("컬럼")` → **GroupBy 객체 반환** - 특정 컬럼으로 **그룹화**하여 집계할 때 사용
  - 예: `df.groupby("Age").mean()` → 나이별 평균 계산

**데이터 필터링**
- `df[조건]` → **DataFrame 반환** - 조건을 만족하는 행만 **필터링**할 때 사용
  - 예: `df[df["Age"] > 30]` → 나이가 30 이상인 행만 선택

````python
print("\n요약 정보")
print(df.info())

print("\n통계 요약")
print(df.describe())

print("\n데이터 크기 (행, 열):", df.shape)

print("\n컬럼 이름들:", df.columns.tolist())

print("\n인덱스:", df.index)

print("\n상위 2개 행 보기")
print(df.head(2))

print("\nScore 평균:", df["Score"].mean())

print("\n나이 기준 정렬")
print(df.sort_values(by="Age"))
````

### TODO 11: `pandas`와 친해지기

pandas.DataFrame은 pandas 라이브러리의 정수입니다. 앞으로도 다양한 데이터를 DataFrame으로 변환할 일이 많습니다. DataFrame은 주어진 예시 말고도 다양한 방법으로 instance를 생성할 수 있습니다.

**요구사항**   
다음은 위에서 생성한 dataframe과 같은 데이터를 다른 방법으로 DataFrame으로 만들어봅시다.
1. `List[Dict]` 형태의 데이터를 만들어봅시다.
2. `List[List]` 형태의 데이터를 통해 만들어봅시다.
3. `numpy.ndarray` 형태의 데이터를 통해 만들어봅시다.

````python
# 기준이 되는 DataFrame
df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "Score": [85.5, 90.3, 78.9]
})

# 1. List[Dict] 형태의 데이터를 통해 DataFrame을 만들어봅시다.
alice = {"Name": "Alice", "Age": 25, "Score": 85.5}
bob = {"Name": "Bob", "Age": 30, "Score": 90.3}
charlie = {"Name": "Charlie", "Age": 35, "Score": 78.9}

df_1 = pd.DataFrame([alice, bob, charlie])
print(df_1)
assert df.equals(df_1), "두 데이터가 다릅니다. column / index / data-type 등을 확인해보세요."

# 2. List[List] 형태의 데이터를 통해 DataFrame을 만들어봅시다.
data = [
    ["Alice", 25, 85.5],
    ["Bob", 30, 90.3],
    ["Charlie", 35, 78.9]
]
columns = ["Name", "Age", "Score"]

df_2 = pd.DataFrame(data, columns=columns)
print(df_2)
assert df.equals(df_2), "두 데이터가 다릅니다. column / index / data-type 등을 확인해보세요."

# 3. numpy.ndarray 형태의 데이터를 통해 DataFrame을 만들어봅시다.
# assertion code에서 에러가 나는 경우, `df`와 `df_3`의 차이를 한 번 살펴봐보세요 :)
names = ["Alice", "Bob", "Charlie"]
nums = np.array([
    [25, 85.5],
    [30, 90.3],
    [35, 78.9]
])

df_3 = pd.DataFrame(nums, columns=["Age", "Score"])
print(df_3)
df_3.insert(0, "Name", names)
print(df_3)
df_3["Age"] = df_3["Age"].astype(int)
assert df.equals(df_3), "두 데이터가 다릅니다. column / index / data-type 등을 확인해보세요."

print("✅ 모든 DataFrame 테스트 통과!")
````
